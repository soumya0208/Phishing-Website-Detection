Droping duplicate data rows

```python
df.drop_duplicates(inplace=True,keep='first')
```

```python
df.shape
```

dropping duplicate columns if any

```python
duplicatecolumns=set()
for x in range(df.shape[1]):
    col1=df.iloc[:,x]
    for y in range(x+1,df.shape[1]):
        col2=df.iloc[:,y]
        if col1.equals(col2):
            duplicatecolumns.add(df.columns.values[y])
print(duplicatecolumns)
```

```python
df.drop(list(duplicatecolumns),inplace=True,axis=1)
```

# Feature Engineering

```python
#creting column function for dropping cloumn
def dropcols(col ,traindf = x_train,valdf = x_val,testdf = x_test):
    traindf.drop(col, axis =1,inplace=True)
    valdf.drop(col,axis=1 , inplace=True)
    testdf.drop(col,axis=1 , inplace=True)
    return traindf,valdf ,testdf
```

```python
x_train.url.value_counts()
```

now if url is a string value and it needs to be label encoded or discarded according to the need . now above observation is telling us that url has too many unique value and if we went for encoding url then it is possible that we ran into multi-label problem .so we are dropping it 

```python
x_train,x_val,x_test = dropcols('url')
```

```python
x_train.shape
```

a column which have to much value repetation with out having any realtionship with target column is known as qusi-constant column we are seraching and droping such column

```python
#checking the quasi-constant columns
sel=VarianceThreshold(threshold=0.1)
sel.fit(x_train)
quarsi_constants=[x for x in x_train.columns if x not in x_train.columns[sel.get_support()]]
quarsi_constants
```

```python
for i in quarsi_constants:
  x_train,x_val,x_test=dropcols(i)
```

now we are going to use much more superiror wrapper methods to make the choice of best features

our first method for wrapper method is sequesntial feature search

```python
#importing the required libraries
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
from sklearn.ensemble import RandomForestRegressor ,RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
```

```python
#selecting the best features using Sequential-Feature-Selector
#searching for best  features set
sfs=SFS(RandomForestClassifier(n_jobs=3),
        k_features='best',
        forward=True,
        floating=False,
        verbose=2,
        scoring='neg_mean_squared_error',
        cv=5
       )
```

```python
#fitting the data
sfs=sfs.fit(np.array(x_train),np.array(y_train))
```

```python
#fetching the selected features using Sequential-Feature-Selector
selected_feat_sfs=x_train.columns[list(sfs.k_feature_idx_)]
selected_feat_sfs=list(selected_feat_sfs)
print(selected_feat_sfs)
print(len(selected_feat_sfs))
```

```python
#creating  dataframes using the selected features
x_model_train=x_train.loc[:,selected_feat_sfs]
x_model_val=x_val.loc[:,selected_feat_sfs]
x_model_test=x_test.loc[:,selected_feat_sfs]
```
